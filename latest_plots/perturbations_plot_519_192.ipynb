```python
#I put this code in my jupyter directory. 
#You probably need to change to dir variable below to point to /data/vxtest/519
#You will also need to change the toutfile to the same location
#The cluster number is ncluster and the set of times
#DTmax is the time interval over which particles left recently which will be followed
#The plot starts at ifile_start
 
import numpy as np
import astropy as ast
import scipy as sc
import matplotlib.pyplot as plt
 

dir = '/data/vxtest/519'
DTmax = 0.4
ncluster = 192
ifile_start = 930
 
plt.rcParams["figure.figsize"] = (16,10)
plt.rcParams.update({'font.size': 14})
 
fig, (ax1) = plt.subplots()
#ax1.set_xlim(-90,90)
#ax1.get_yaxis().set_visible(False)
#ax1.set_xlim(-180,180)
#ax1.grid(True)
#ygrid = np.arange(200)*30
#ax1.set_yticks(ygrid)
ax1.set_xlabel('Time [Gyr]',size='x-large')
ax1.set_ylabel('L - <L> [kpc km/sec]',size='x-large')
 
nskip=1
nplot = int((1330-ifile_start)/nskip + 1)
print("nplot ",nplot)
ifile = ifile_start
toutfile =   dir +'/TOUT' + f'{ifile:04}'
thead = np.fromfile(toutfile,dtype='double',count=1)
Tend = thead[0]
print(ifile,Tend)
nnn = np.fromfile(toutfile,dtype='int32',count=3,offset=8)
nstart = nnn[0]
ntc = nnn[1]
nts = nnn[2]
Toffset = 14.1 - Tend
print(Tend,nstart,ntc,nts,Toffset)
noff = 8 + 3*4
Nco = np.zeros(ntc,dtype='int32')
outid = []
ctout = []
for i in range(ntc):
    Nco[i] = np.fromfile(toutfile,dtype='int32',count=1,offset=noff)
    noff += 4
    readoutN = np.fromfile(toutfile,dtype='int32',count=Nco[i],offset=noff)
    outid.append(readoutN)
    noff += 4*Nco[i]
    readoutT= np.fromfile(toutfile,dtype='double',count=Nco[i],offset=noff)
    ctout.append(readoutT)
    noff += 8*Nco[i]

cplot = ctout[ncluster] < DTmax
print("Cluster",ncluster,Tend,DTmax,np.sum(cplot))
ptime = []
Ltime = []
for i in range(nplot):
    ifile = ifile_start + nskip*i
    filename =   dir+'/snapshot_' + f'{ifile:04}'
    hd = np.fromfile(filename,dtype='int32',count=7)
    nd = hd[2]
    ns = hd[5]
    nc = hd[3]
    nb = hd[4]
    head = np.fromfile(filename,dtype='double',count=7,offset=4+6*4)
    time=head[6]
 
    start_d = 0
    start_c = nd
    start_s = nd + nc + nb
    start_b = nd + nc
    block = nd + nc + ns + nb
 
    skipx = 256 + 3*4 + start_b*3*8
    skipv = 256 + 5*4 + (block*3 + start_b*3)*8
    rc = np.fromfile(filename,dtype='double',count=3,offset=skipx)
    vc = np.fromfile(filename,dtype='double',count=3,offset=skipv)
    xh = rc[0]
    yh = rc[1]
    zh = rc[2]
    vxh = vc[0]
    vyh = vc[1]
    vzh = vc[2]
 
    skipx = 256 + 3*4 + start_c*3*8
    skipv = 256 + 5*4 + (block*3 + start_c*3)*8
    skipi = 256 + 7*4 + block*3*2*8 + start_c*4
    rc = np.fromfile(filename,dtype='double',count=3*nc,offset=skipx)
    vc = np.fromfile(filename,dtype='double',count=3*nc,offset=skipv)
    idc = np.fromfile(filename,dtype='uint32',count=nc,offset=skipi)
    jxc = rc[0::3] - xh
    jyc = rc[1::3] - yh
    jzc = rc[2::3] - zh
 
    jvxc = vc[0::3] - vxh
    jvyc = vc[1::3] - vyh
    jvzc = vc[2::3] - vzh
 
    idc -= nd
    ki = np.zeros(nc,dtype='int32')
    for j in range(nc):
        ix = idc[j]
        ki[ix] = j
 
    rh3 = [xh,yh,zh]
    vh3 = [vxh,vyh,vzh]
 
    #centers unsorted
    jrc3 = np.reshape(rc,(nc,3)) - rh3
    jvc3 = np.reshape(vc,(nc,3)) - vh3
 
    rc3 = jrc3[ki]
    vc3 = jvc3[ki]
 
    #stars
    skipx = 256 + 3*4 + start_s*3*8
    skipv = 256 + 5*4 + block*3*8 + start_s*3*8 #skip over r, and dm+c v's
    skipi = 256 + 7*4 + block*3*2*8 + start_s*4 #skip over r and v and dm+c id
    skipc = 256 + 9*4 + block*(2*3*8+4) + start_s*4 #skip over r, v and id and dm+c ic
 
    rs= np.fromfile(filename,dtype='double',count=3*ns,offset=skipx)
    vs = np.fromfile(filename,dtype='double',count=3*ns,offset=skipv)
    ids = np.fromfile(filename,dtype='uint32',count=ns,offset=skipi)
    jics = np.fromfile(filename,dtype='uint32',count=ns,offset=skipc)
 
    jrs3 = np.reshape(rs,(ns,3)) - rh3
    jvs3 = np.reshape(vs,(ns,3)) - vh3
 
    ids -= nd + nc + nb
    ki = np.zeros(ns,dtype='int32')
    for j in range(ns):
        ix = ids[j]
        ki[ix] = j
 
    rs3 = jrs3[ki]
    vs3 = jvs3[ki]
    ics = jics[ki]
 
    dsm = np.linalg.norm(rs3,axis=1)
    vsm  =np.linalg.norm(vs3,axis=1)
 
    Ls3 = np.cross(rs3,vs3)
    Lsm = np.linalg.norm(Ls3,axis=1)
    vts = Lsm/dsm
    vrs = (rs3*vs3).sum(axis=1)/dsm
 
    dcm = np.linalg.norm(rc3,axis=1)
    vcm  =np.linalg.norm(vc3,axis=1)
 
    Lc3 = np.cross(rc3,vc3)
    Lcm = np.linalg.norm(Lc3,axis=1)
    vtc = Lcm/dcm
    vrc =  (rc3*vc3).sum(axis=1)/dcm
 
    lon = np.arctan2(rs3[:,1],rs3[:,0])
    lat = np.arcsin(rs3[:,2]/dsm)
    Lsz = np.abs(Ls3[:,2])
 
    lonc = np.arctan2(rc3[:,1],rc3[:,0])
    latc = np.arcsin(rc3[:,2]/dcm)
    Lcz = np.abs(Lc3[:,2])
    tempYc3 = -np.cross(rc3,np.cross(rc3,vc3))
    normYc3 = np.sqrt((tempYc3 ** 2).sum(-1))[..., np.newaxis]
    Yc3 = tempYc3 / normYc3
 
    rotx = np.sum(rc3[ics] * rs3,axis=1)/dcm[ics]
    roty = np.sum(Yc3[ics] * rs3,axis=1)
    rotz = np.sum(Lc3[ics] * rs3,axis=1)/Lcm[ics]
    rotlen = np.sqrt(rotx**2 + roty**2 + rotz**2)
    rotlon = np.arctan2(roty,rotx)
    rotlat = np.arcsin(rotz/rotlen)
    if i == 0:
        inLs30 = Ls3[outid[ncluster]]
   
    rotvx = np.sum(rc3[ics] * vs3,axis=1)/dcm[ics]
    rotvy = np.sum(Yc3[ics] * vs3,axis=1)
    rotvz = np.sum(Lc3[ics] * vs3,axis=1)/Lcm[ics]
 
    rotcx = np.sum(rc3 * rc3,axis=1)/dcm
    rotcy = np.sum(Yc3 * rc3,axis=1)
    rotcz = np.sum(Lc3 * rc3,axis=1)/Lcm
 
    rotclen = np.sqrt(rotcx**2 + rotcy**2 + rotcz**2)
    rotclon = np.arctan2(rotcy,rotcx)
    rotclat = np.arcsin(rotcz/rotclen)
 
    inLs3 = Ls3[outid[ncluster]]
    Lsav = np.mean(np.linalg.norm(inLs3,axis=1))
    dLs = inLs3[cplot] - inLs30[cplot] #- Lc3[ncluster]
    dLm = np.linalg.norm(dLs,axis=1)
    dLav = np.median(dLm)
    dLst = np.var(dLm)
    dLm -= dLav
    atime = np.full_like(dLm,time)
    ptime.append(atime)
    Ltime.append(dLm)
    ax1.scatter(time,Lsav/40,color='black',s=2) #L scaling
    indsm = dsm[outid[ncluster]]
    dmedian = np.median(indsm[cplot])
    ax1.scatter(time,(dmedian-200),color='red',s=2) #L scaling
    print(i,ifile,f'{time:.2f} {dLav:.1f} {np.sqrt(dLst):.1f} {Lsav:.1f} {Lcm[ncluster]:.1f} {dmedian:.1f}')
    #plotoff += plotinc

ax1.set_xlim(Tend-DTmax/2,14.2)
ax1.set_ylim(-200,200)
 
for ip in range(len(dLm)):
    plott = np.zeros(nplot)
    plotL = np.zeros(nplot)
    for it in range(nplot):
        plott[it] =ptime[it][ip]
        plotL[it] =Ltime[it][ip]
    ax1.plot(plott,plotL,linewidth=1)#,color='black')

ax1.annotate('Nc '+ f'{ncluster:04}' + " DT "+f'{DTmax:3.1f}' + ' Lscale=40',xy=(Tend,180))#,size='large')
#plt.savefig("DLT" + f'{ifile_start:04}'+"_" f'{ncluster:04}'+ '.jpeg',dpi=300)
plt.savefig("DLT" + f'{ifile_start:04}'+"_" f'{ncluster:04}'+ '.pdf',dpi=300)
```
