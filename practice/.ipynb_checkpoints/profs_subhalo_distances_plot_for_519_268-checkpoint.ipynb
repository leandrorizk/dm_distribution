```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
 
dir = '/data/vxtest/519'
plt.rcParams["figure.figsize"] = (16,10)
plt.rcParams.update({'font.size': 14})
 
fig, (ax1) = plt.subplots()
#ax1.set_xlim(-90,90)
#ax1.get_yaxis().set_visible(False)
#ax1.set_xlim(-180,180)
#ax1.grid(True)
#ygrid = np.arange(200)*30
#ax1.set_yticks(ygrid)
ax1.set_xlabel('Time [Gyr]',size='x-large')
ax1.set_ylabel('distance [kpc]',size='x-large')
 
DTmax = 0.4
ncluster = 268
 
nskip=1
ifile_start = 930
ifile_dstart = 1150-80 #
ifile_dend = 1200-80 #
ptbeg = (ifile_dstart+80)*0.01
nplot = int((ifile_dend-ifile_dstart)/nskip + 1)
print("nplot ",nplot)
ifile = ifile_start
toutfile =   dir +'/TOUT' + f'{ifile:04}'
thead = np.fromfile(toutfile,dtype='double',count=1)
Tend = thead[0]
print(ifile,Tend)
nnn = np.fromfile(toutfile,dtype='int32',count=3,offset=8)
nstart = nnn[0]
ntc = nnn[1]
nts = nnn[2]
Toffset = 14.1 - Tend
print(Tend,nstart,ntc,nts,Toffset)
noff = 8 + 3*4
Nco = np.zeros(ntc,dtype='int32')
outid = []
ctout = []
for i in range(ntc):
    Nco[i] = np.fromfile(toutfile,dtype='int32',count=1,offset=noff)
    noff += 4
    readoutN = np.fromfile(toutfile,dtype='int32',count=Nco[i],offset=noff)
    outid.append(readoutN)
    noff += 4*Nco[i]
    readoutT= np.fromfile(toutfile,dtype='double',count=Nco[i],offset=noff)
    ctout.append(readoutT)
    noff += 8*Nco[i]

cplot = ctout[ncluster] < DTmax
print("Cluster",ncluster,Tend,DTmax,np.sum(cplot))
ptime = []
for i in (range(nplot)):
    ifile = ifile_dstart + nskip*i
    filename =   dir+'/snapshot_' + f'{ifile:04}'
    hd = np.fromfile(filename,dtype='int32',count=7)
    nd = hd[2]
    ns = hd[5]
    nc = hd[3]
    nb = hd[4]
    head = np.fromfile(filename,dtype='double',count=7,offset=4+6*4)
    time=head[6]
 
    start_d = 0
    start_c = nd
    start_s = nd + nc + nb
    start_b = nd + nc
    block = nd + nc + ns + nb
 
    skipx = 256 + 3*4 + start_b*3*8 #bulge particle
    skipv = 256 + 5*4 + (block*3 + start_b*3)*8
    rc = np.fromfile(filename,dtype='double',count=3,offset=skipx)
    vc = np.fromfile(filename,dtype='double',count=3,offset=skipv)
    xh = rc[0]
    yh = rc[1]
    zh = rc[2]
    vxh = vc[0]
    vyh = vc[1]
    vzh = vc[2]
 
    skipx = 256 + 3*4 + start_c*3*8
    skipv = 256 + 5*4 + (block*3 + start_c*3)*8
    skipi = 256 + 7*4 + block*3*2*8 + start_c*4
    rc = np.fromfile(filename,dtype='double',count=3*nc,offset=skipx)
    vc = np.fromfile(filename,dtype='double',count=3*nc,offset=skipv)
    idc = np.fromfile(filename,dtype='uint32',count=nc,offset=skipi)
    jxc = rc[0::3] - xh
    jyc = rc[1::3] - yh
    jzc = rc[2::3] - zh
 
    jvxc = vc[0::3] - vxh
    jvyc = vc[1::3] - vyh
    jvzc = vc[2::3] - vzh
 
    idc -= nd
    ki = np.zeros(nc,dtype='int32')
    for j in range(nc):
        ix = idc[j]
        ki[ix] = j
 
    rh3 = [xh,yh,zh]
    vh3 = [vxh,vyh,vzh]
 
    #centers unsorted
    jrc3 = np.reshape(rc,(nc,3)) - rh3
    jvc3 = np.reshape(vc,(nc,3)) - vh3
 
    rc3 = jrc3[ki]
    vc3 = jvc3[ki]
 
    #stars
    skipx = 256 + 3*4 + start_s*3*8
    skipv = 256 + 5*4 + block*3*8 + start_s*3*8 #skip over r, and dm+c v's
    skipi = 256 + 7*4 + block*3*2*8 + start_s*4 #skip over r and v and dm+c id
    skipc = 256 + 9*4 + block*(2*3*8+4) + start_s*4 #skip over r, v and id and dm+c ic
 
    rs= np.fromfile(filename,dtype='double',count=3*ns,offset=skipx)
    vs = np.fromfile(filename,dtype='double',count=3*ns,offset=skipv)
    ids = np.fromfile(filename,dtype='uint32',count=ns,offset=skipi)
    jics = np.fromfile(filename,dtype='uint32',count=ns,offset=skipc)
 
    jrs3 = np.reshape(rs,(ns,3)) - rh3
    jvs3 = np.reshape(vs,(ns,3)) - vh3
 
    ids -= nd + nc + nb
    ki = np.zeros(ns,dtype='int32')
    for j in range(ns):
        ix = ids[j]
        ki[ix] = j
 
    rs3 = jrs3[ki]
    vs3 = jvs3[ki]
    ics = jics[ki]
 
    inx = rs3[outid[ncluster],0]
    iny = rs3[outid[ncluster],1]
    inz = rs3[outid[ncluster],2]
    minx = np.median(inx[cplot])
    miny = np.median(iny[cplot])
    minz = np.median(inz[cplot])
    stdx = np.std(inx[cplot])
    stdy = np.std(iny[cplot])
    stdz = np.std(inz[cplot])
    invx = vs3[outid[ncluster],0]
    invy = vs3[outid[ncluster],1]
    invz = vs3[outid[ncluster],2]
    minvx = np.median(invx[cplot])
    minvy = np.median(invy[cplot])
    minvz = np.median(invz[cplot])
 
   
    haloname =  f'{dir}'+'/halos/h' + f'{ifile}'+ '.z0.000.AHF_halos'
    dh = np.loadtxt(haloname)
    halom = dh[:,3]
    halo_min = halom > 1.e7
    halox = dh[:,5] - xh
    haloy = dh[:,6] - yh
    haloz = dh[:,7] - zh
    halovx = dh[:,8] - vxh
    halovy = dh[:,9] - vyh
    halovz = dh[:,10] - vzh
    hx = halox[halo_min]
    hy = haloy[halo_min]
    hz = haloz[halo_min]
    hvx = halovx[halo_min]
    hvy = halovy[halo_min]
    hvz = halovz[halo_min]
    hm = halom[halo_min]
   
    dx = hx - minx
    dy = hy - miny
    dz = hz - minz
    dh = np.sqrt(dx**2 + dy**2 +dz**2)
    dvx = hvx-minvx
    dvy = hvy-minvy
    dvz = hvz-minvz
   
    dvh = np.sqrt(dvx**2 + dvy**2 + dvz**2)
    dvph = (dx*dvx + dy*dvy + dz*dvz)/dh
    loghm = np.log10(hm)
    psize = 15*(loghm-6)
    ptime = np.full_like(dh,time)
    mprint1 = dh < 15
    mprint2 = loghm > 7
    mprint = mprint1 & mprint2
    print('T',f'{time:.2f}',f'{minx:.1f}',f'{miny:.1f}', f'{minz:.1f}',f'{minvx:.1f}',f'{minvy:.1f}', f'{minvz:.1f}' )
    dhp = dh[mprint]
    dvhp = dvh[mprint]
    dvphp = dvph[mprint]
    hmp = loghm[mprint]
    if np.sum(mprint) >0:
        np.set_printoptions(precision=1)
        print('D',dhp)
        print('V',dvhp)
        print('Vp',dvphp)
        np.set_printoptions(precision=3)
        print(hmp)
    ax1.scatter(ptime,dh,color='black',s=psize)
    ax1.scatter(time,np.sqrt(stdx**2+stdy**2+stdz**2),color='red')

ax1.set_xlim(0.01*(ifile_dstart+80), 0.01*(ifile_dend+80))
ax1.set_ylim(0,25)
 
#ax1.annotate('Nc '+ f'{ncluster:04}' + " DT "+f'{DTmax:3.1f}',xy=(ptbeg+0.05,5))#,size='large')
#plt.savefig(dir +"/DLT" + f'{ifile_start:04}'+"_" f'{ncluster:04}'+ '.jpeg',dpi=300)

# plt.savefig(dir +"/DHaloT" + f'{ifile_start:04}'+"_" f'{ncluster:04}'+ '.jpeg',dpi=300)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
 
dir = '/data/vxtest/519'
plt.rcParams["figure.figsize"] = (16,10)
plt.rcParams.update({'font.size': 14})
 
fig, (ax1) = plt.subplots()
#ax1.set_xlim(-90,90)
#ax1.get_yaxis().set_visible(False)
#ax1.set_xlim(-180,180)
#ax1.grid(True)
#ygrid = np.arange(200)*30
#ax1.set_yticks(ygrid)
ax1.set_xlabel('Time [Gyr]',size='x-large')
ax1.set_ylabel('distance [kpc]',size='x-large')
 
DTmax = 0.4
ncluster = 268
 
nskip=1
ifile_start = 930
ifile_dstart = 1150-80 #
ifile_dend = 1200-80 #
ptbeg = (ifile_dstart+80)*0.01
nplot = int((ifile_dend-ifile_dstart)/nskip + 1)
print("nplot ",nplot)
ifile = ifile_start
toutfile =   dir +'/TOUT' + f'{ifile:04}'
thead = np.fromfile(toutfile,dtype='double',count=1)
Tend = thead[0]
print(ifile,Tend)
nnn = np.fromfile(toutfile,dtype='int32',count=3,offset=8)
nstart = nnn[0]
ntc = nnn[1]
nts = nnn[2]
Toffset = 14.1 - Tend
print(Tend,nstart,ntc,nts,Toffset)
noff = 8 + 3*4
Nco = np.zeros(ntc,dtype='int32')
outid = []
ctout = []
for i in range(ntc):
    Nco[i] = np.fromfile(toutfile,dtype='int32',count=1,offset=noff)
    noff += 4
    readoutN = np.fromfile(toutfile,dtype='int32',count=Nco[i],offset=noff)
    outid.append(readoutN)
    noff += 4*Nco[i]
    readoutT= np.fromfile(toutfile,dtype='double',count=Nco[i],offset=noff)
    ctout.append(readoutT)
    noff += 8*Nco[i]

cplot = ctout[ncluster] < DTmax
print("Cluster",ncluster,Tend,DTmax,np.sum(cplot))
ptime = []
for i in (range(nplot)):
    ifile = ifile_dstart + nskip*i
    filename =   dir+'/snapshot_' + f'{ifile:04}'
    hd = np.fromfile(filename,dtype='int32',count=7)
    nd = hd[2]
    ns = hd[5]
    nc = hd[3]
    nb = hd[4]
    head = np.fromfile(filename,dtype='double',count=7,offset=4+6*4)
    time=head[6]
 
    start_d = 0
    start_c = nd
    start_s = nd + nc + nb
    start_b = nd + nc
    block = nd + nc + ns + nb
 
    skipx = 256 + 3*4 + start_b*3*8 #bulge particle
    skipv = 256 + 5*4 + (block*3 + start_b*3)*8
    rc = np.fromfile(filename,dtype='double',count=3,offset=skipx)
    vc = np.fromfile(filename,dtype='double',count=3,offset=skipv)
    xh = rc[0]
    yh = rc[1]
    zh = rc[2]
    vxh = vc[0]
    vyh = vc[1]
    vzh = vc[2]
 
    skipx = 256 + 3*4 + start_c*3*8
    skipv = 256 + 5*4 + (block*3 + start_c*3)*8
    skipi = 256 + 7*4 + block*3*2*8 + start_c*4
    rc = np.fromfile(filename,dtype='double',count=3*nc,offset=skipx)
    vc = np.fromfile(filename,dtype='double',count=3*nc,offset=skipv)
    idc = np.fromfile(filename,dtype='uint32',count=nc,offset=skipi)
    jxc = rc[0::3] - xh
    jyc = rc[1::3] - yh
    jzc = rc[2::3] - zh
 
    jvxc = vc[0::3] - vxh
    jvyc = vc[1::3] - vyh
    jvzc = vc[2::3] - vzh
 
    idc -= nd
    ki = np.zeros(nc,dtype='int32')
    for j in range(nc):
        ix = idc[j]
        ki[ix] = j
 
    rh3 = [xh,yh,zh]
    vh3 = [vxh,vyh,vzh]
 
    #centers unsorted
    jrc3 = np.reshape(rc,(nc,3)) - rh3
    jvc3 = np.reshape(vc,(nc,3)) - vh3
 
    rc3 = jrc3[ki]
    vc3 = jvc3[ki]
 
    #stars
    skipx = 256 + 3*4 + start_s*3*8
    skipv = 256 + 5*4 + block*3*8 + start_s*3*8 #skip over r, and dm+c v's
    skipi = 256 + 7*4 + block*3*2*8 + start_s*4 #skip over r and v and dm+c id
    skipc = 256 + 9*4 + block*(2*3*8+4) + start_s*4 #skip over r, v and id and dm+c ic
 
    rs= np.fromfile(filename,dtype='double',count=3*ns,offset=skipx)
    vs = np.fromfile(filename,dtype='double',count=3*ns,offset=skipv)
    ids = np.fromfile(filename,dtype='uint32',count=ns,offset=skipi)
    jics = np.fromfile(filename,dtype='uint32',count=ns,offset=skipc)
 
    jrs3 = np.reshape(rs,(ns,3)) - rh3
    jvs3 = np.reshape(vs,(ns,3)) - vh3
 
    ids -= nd + nc + nb
    ki = np.zeros(ns,dtype='int32')
    for j in range(ns):
        ix = ids[j]
        ki[ix] = j
 
    rs3 = jrs3[ki]
    vs3 = jvs3[ki]
    ics = jics[ki]
 
    inx = rs3[outid[ncluster],0]
    iny = rs3[outid[ncluster],1]
    inz = rs3[outid[ncluster],2]
    minx = np.median(inx[cplot])
    miny = np.median(iny[cplot])
    minz = np.median(inz[cplot])
    stdx = np.std(inx[cplot])
    stdy = np.std(iny[cplot])
    stdz = np.std(inz[cplot])
    invx = vs3[outid[ncluster],0]
    invy = vs3[outid[ncluster],1]
    invz = vs3[outid[ncluster],2]
    minvx = np.median(invx[cplot])
    minvy = np.median(invy[cplot])
    minvz = np.median(invz[cplot])
 
   
    haloname =  f'{dir}'+'/halos/h' + f'{ifile}'+ '.z0.000.AHF_halos'
    dh = np.loadtxt(haloname)
    halom = dh[:,3]
    halo_min = halom > 1.e7
    halox = dh[:,5] - xh
    haloy = dh[:,6] - yh
    haloz = dh[:,7] - zh
    halovx = dh[:,8] - vxh
    halovy = dh[:,9] - vyh
    halovz = dh[:,10] - vzh
    hx = halox[halo_min]
    hy = haloy[halo_min]
    hz = haloz[halo_min]
    hvx = halovx[halo_min]
    hvy = halovy[halo_min]
    hvz = halovz[halo_min]
    hm = halom[halo_min]
   
    dx = hx - minx
    dy = hy - miny
    dz = hz - minz
    dh = np.sqrt(dx**2 + dy**2 +dz**2)
    dvx = hvx-minvx
    dvy = hvy-minvy
    dvz = hvz-minvz
   
    dvh = np.sqrt(dvx**2 + dvy**2 + dvz**2)
    dvph = (dx*dvx + dy*dvy + dz*dvz)/dh
    loghm = np.log10(hm)
    psize = 15*(loghm-6)
    ptime = np.full_like(dh,time)
    mprint1 = dh < 15
    mprint2 = loghm > 7
    mprint = mprint1 & mprint2
    print('T',f'{time:.2f}',f'{minx:.1f}',f'{miny:.1f}', f'{minz:.1f}',f'{minvx:.1f}',f'{minvy:.1f}', f'{minvz:.1f}' )
    dhp = dh[mprint]
    dvhp = dvh[mprint]
    dvphp = dvph[mprint]
    hmp = loghm[mprint]
    if np.sum(mprint) >0:
        np.set_printoptions(precision=1)
        print('D',dhp)
        print('V',dvhp)
        print('Vp',dvphp)
        np.set_printoptions(precision=3)
        print(hmp)
    ax1.scatter(ptime,dh,color='black',s=psize)
    ax1.scatter(time,np.sqrt(stdx**2+stdy**2+stdz**2),color='red')

ax1.set_xlim(0.01*(ifile_dstart+80), 0.01*(ifile_dend+80))
ax1.set_ylim(0,25)
 
#ax1.annotate('Nc '+ f'{ncluster:04}' + " DT "+f'{DTmax:3.1f}',xy=(ptbeg+0.05,5))#,size='large')
#plt.savefig(dir +"/DLT" + f'{ifile_start:04}'+"_" f'{ncluster:04}'+ '.jpeg',dpi=300)

# plt.savefig(dir +"/DHaloT" + f'{ifile_start:04}'+"_" f'{ncluster:04}'+ '.jpeg',dpi=300)
```
