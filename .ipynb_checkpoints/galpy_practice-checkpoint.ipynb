```python
import galpy
import numpy as np
import astropy.units as u
import astropy.constants as c
```

```python
m = 10e8 * u.solMass
v = 200 * u.km / u.s
b = 1 * u.kpc
r = 20 * u.kpc
```

```python
delta_v = 2 * c.G * m / (b * v)
```

```python
delta_v = delta_v.to(u.km / u.s)
delta_v
```

```python
delta_ang_mome = r * delta_v
delta_ang_mome
```
